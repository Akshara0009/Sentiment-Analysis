**DATA PREPROCESSING**

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk import download
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(raw_text, flag):
    #convert the list to a string
    x = str(raw_text)
    # Removing special characters and digits
    sentence = re.sub('[^a-zA-Z]',' ',x)
    # change sentence to lower case
    sentence = sentence.lower()
    # tokenize into words
    tokens = sentence.split()
    # remove stop words
    clean_tokens = [t for t in tokens if not t in stopwords.words("english")]
    # Stemming/Lemmatization
    if(flag == 'lemma'):
        clean_tokens = [lemmatizer.lemmatize(word) for word in clean_tokens]
    return pd.Series([" ".join(clean_tokens), len(clean_tokens)])
```

```python
import nltk
nltk.download("stopwords")
nltk.download("wordnet")
```

**CONVERTING TEXT TO NUMERICAL VECTORS**

```python
# Apply preprocessing to the ReviewText column
X_train[["processed_ReviewText", "ReviewText_word_count"]] = X_train["ReviewText"].apply(lambda x: preprocess(x, "lemma"))
print(X_train.head())
```

```python
# Apply preprocessing to the ReviewText column in X_test
X_test[["processed_ReviewText", "ReviewText_word_count"]] = X_test["ReviewText"].apply(lambda x: preprocess(x, "lemma"))
print(X_test.head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vocab = CountVectorizer(max_features=5000,min_df=5, max_df=0.7)
```

```python
X_train_transformed = vocab.fit_transform(X_train["processed_ReviewText"])
```

```python
X_test_transformed = vocab.transform(X_test["processed_ReviewText"])
```

**MODEL BUILDING**

```python
from sklearn.metrics import accuracy_score, classification_report
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train_transformed, y_train)
```

```python
y_test_pred = LR.predict(X_test_transformed)
```

```python
# Evaluation Model
print(accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(X_train_transformed, y_train)
```

```python
y_test_pred = DT.predict(X_test_transformed)
```

```python
# Evaluation Model
print(accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

*Lower performance of classes 2,3 and 4.*

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
RF.fit(X_train_transformed, y_train)
```

```python
y_test_pred = RF.predict(X_test_transformed)
```

```python
# Evaluation Model
print(accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```
